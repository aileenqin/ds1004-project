```python
import numpy as np
f = pd.read_csv("C:/Users/Aileen/Desktop/results/PREM_TYP_DESC.txt", sep = '\t',header = None)
plt.figure(figsize=(20,10))
loc = []
freq = []
for i in f[0]:
    a,b = i.split(',')
    loc.append(a)
    freq.append(int(b))
indices = np.argsort(freq)[::-1]
plt.bar(range(len(loc)), [freq[i] for i in indices], color = 'lightgreen')
indice = [loc[i] for i in indices]
plt.xticks(range(len(loc)), indice)
plt.xticks(rotation=90)
plt.title("Locations Plot ")
plt.xlabel("Location")
plt.ylabel("Frequency")

plt.savefig('C:/Users/Aileen/Desktop/results/PREM_TYP_DESC.png')
plt.show()
```

```python
#ADDR_PCT_CD

labels = ['VALID', 'NULL']
sizes = [5100841, 390]

colors = ['lightblue','yellow']
explode = (0.1, 0)  # explode 1st slice
 
    
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct


# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct=make_autopct(sizes), shadow=True,startangle=140)
 
plt.axis('equal')
plt.title('ADDR_PCT_CD')
# plt.savefig('ADDR_PCT_CD.png')
plt.show()
```
